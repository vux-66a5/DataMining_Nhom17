# Import thư viện và đọc file dữ liệu

```python
import json
import random

# Đọc dữ liệu từ file output.json
with open('output.json', 'r', encoding='utf-8') as file:
    data = json.load(file)
    
transactions = []
```

# Các danh mục

```python

banh_burger = [item['name'] for item in data['Bánh burger']]
banh_trang_mieng = [item['name'] for item in data['Bánh tráng miệng']]
combos = [item['name'] for item in data['Combo']]
ga_ran = [item['name'] for item in data['Gà rán']]
kem = [item['name'] for item in data['Kem']]
mon_an_nhe = [item['name'] for item in data['Món ăn nhẹ']]
thuc_uong_lanh = [item['name'] for item in data['Thức uống lạnh']]
thuc_uong_nong = [item['name'] for item in data['Thức uống nóng']]
```

# Sinh các giao dịch ( đơn hàng ) ngẫu nhiên theo tỉ lệ: 
10% Gà rán + Thức Uống lạnh + Kem
10% Burger + Thức Uống lạnh + Kem
30% Combo + Kem
10% Thức ăn nhẹ + thức uống lạnh
5% bánh tráng miệng + thức uống nóng
5% bánh tráng miệng + thức uống lạnh
20% Gà rán + món ăn nhẹ + kem + thức uống lạnh
10% Burger + Món Ăn Nhẹ + Thức Uống lạnh

```python
# Tạo giao dịch cho 10% Burger + Thức Uống lạnh + Kem
for _ in range(10000):
     transactions.append([random.choice(banh_burger), random.choice(thuc_uong_lanh),random.choice(kem)])
# Tạo giao dịch cho 10% Gà rán + Thức Uống lạnh + Kem
for _ in range(10000):
     transactions.append([random.choice(ga_ran), random.choice(thuc_uong_lanh),random.choice(kem)])     
# Tạo giao dịch cho 30% Combo + kem
for _ in range(30000):
    transactions.append([random.choice(combos), random.choice(kem)]) 
# Tạo giao dịch cho 10% Thức ăn nhẹ + thức uống lạnh
for _ in range(10000):
    transactions.append([random.choice(mon_an_nhe), random.choice(thuc_uong_lanh)])    
# Tạo giao dịch cho 5% Bánh tráng miệng + thức uống nóng
for _ in range(5000):
    transactions.append([random.choice(banh_trang_mieng), random.choice(thuc_uong_nong)])    
# Tạo giao dịch cho 5% Bánh tráng miệng + thức uống lạnh
for _ in range(5000):
    transactions.append([random.choice(banh_trang_mieng), random.choice(thuc_uong_lanh)])
# Tạo giao dịch cho 20% Gà rán + món ăn nhẹ + kem + thức uống lạnh
for _ in range(20000):
    transactions.append([random.choice(ga_ran), random.choice(mon_an_nhe), random.choice(kem), random.choice(thuc_uong_lanh)])     
# Tạo giao dịch cho 10% Bánh burger + món ăn nhẹ + thức uống lạnh
for _ in range(10000):
    transactions.append([random.choice(banh_burger), random.choice(mon_an_nhe), random.choice(thuc_uong_lanh)])
    
```

# Lưu dữ liệu giao dịch vào file

```python

with open('transactions.json', 'a', encoding='utf-8') as file:
    json.dump(transactions, file, ensure_ascii=False, indent=4)
```

# Load file giao dịch đã sinh để xử lí

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

# Đọc dữ liệu giao dịch
with open('transactions.json', 'r', encoding='utf-8') as file:
    transactions = json.load(file)
```

# Chuyển đổi dữ liệu giao dịch sang định dạng DataFrame


```python
df = pd.DataFrame(transactions)
```

```python
# Chuyển đổi dữ liệu thành định dạng phù hợp cho thuật toán Apriori
# basket = df.stack().groupby(level=0).apply(lambda x: pd.Series(1, index=x)).unstack().fillna(0)
# Convert the transaction data into a binary format suitable for Apriori algorithm
basket = pd.get_dummies(df.stack()).groupby(level=0).max()
```

```python
print(type(basket))
# print(tabulate(basket, headers = 'keys', tablefmt = 'github'))
display(basket)
```

```python
# Apply Apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(basket, min_support=0.05, use_colnames=True)
```

```python
# Generate association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.4)
```

```python
# Display frequent itemsets and association rules
print("Frequent Itemsets:")
print(frequent_itemsets)

print("\nAssociation Rules:")
print(rules)
```

```python
# Áp dụng thuật toán Apriori
frequent_itemsets = apriori(basket, min_support=0.005, use_colnames=True)
# rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
print(frequent_itemsets)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.4)
```

```python
# Định dạng bảng đẹp với print và vòng lặp
for index, row in rules.iterrows():
    print(f"Rule {index + 1}:")
    print(f"  Antecedent: {', '.join(row['antecedents'])}")
    print(f"  Consequent: {', '.join(row['consequents'])}")
    print(f"  Antecedent support: {row['antecedent support']:.3f}")
    print(f"  Consequent support: {row['consequent support']:.3f}")
    print(f"  Support: {row['support']:.3f}")
    print(f"  Confidence: {row['confidence']:.3f}")
    print(f"  Lift: {row['lift']:.3f}")
    print(f"  Leverage: {row['leverage']:.3f}")
    print(f"  Conviction: {row['conviction']:.3f}")
    print(f"  Zhang's Metric: {row['zhangs_metric']:.3f}")
    print("-" * 50)
```

```python
import json

# Convert sets to lists in the DataFrame
rules['antecedents'] = rules['antecedents'].apply(lambda x: list(x))
rules['consequents'] = rules['consequents'].apply(lambda x: list(x))

# Convert DataFrame to list of dictionaries
rules_list = rules.to_dict(orient='records')

# Specify the output file path
output_file = 'rules.json'

# Write rules to a JSON file
with open(output_file, 'w', encoding='utf-8') as file:
    json.dump(rules_list, file, ensure_ascii=False, indent=4)

print(f"Rules have been exported to the file: {output_file}")
```

```python
import json

# Đọc dữ liệu từ file rules.json
with open('rules.json', 'r', encoding='utf-8') as file:
    rules = json.load(file)

# Hàm để tìm các đề xuất dựa trên sản phẩm đầu vào
def find_recommendations(product):
    recommendations = []
    for rule in rules:
        antecedents = rule['antecedents']
        consequents = rule['consequents']
        if product in antecedents:
            recommendations.append(consequents)
    return recommendations

# Hỏi người dùng sản phẩm muốn tìm đề xuất
product_input = input("Nhập tên sản phẩm muốn tìm đề xuất: ")

# Tìm các đề xuất dựa trên sản phẩm đầu vào
recommendations = find_recommendations(product_input)

# In kết quả
if recommendations:
    print(f"Các đề xuất cho sản phẩm '{product_input}':")
    for recommendation in recommendations:
        print(", ".join(recommendation))
else:
    print(f"Không tìm thấy đề xuất cho sản phẩm '{product_input}'.")
```
